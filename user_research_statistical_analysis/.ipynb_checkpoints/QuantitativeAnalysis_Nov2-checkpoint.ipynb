```python
import numpy as np
%matplotlib inline
import matplotlib
matplotlib.rcParams.update({'font.size':14})
import matplotlib.pyplot as plt 
counterPlot = 1
amountPlot = 3
plt.rcParams["figure.figsize"]=[10,7*amountPlot]
colorF2F = '#8BA6FF'
colorCSCW = '#CC6C52'

# question 1
dataQuestion1 = np.genfromtxt('Question1.csv', dtype=int, delimiter=',', skip_header=1,
	names=True)
dataAction = [[0 for i in range(8)] for j in range(2)]
dataRefContent = [[0 for k in range(5)] for j in range(2)]
dataRefPurpose = [[0 for k in range(5)] for j in range(2)]
for j in range(2):
	for i in range(8):
		# dataXXX[0]: F2F; dataXXX[1]:CSCW
		dataAction[j][i] = dataQuestion1[3+j*4][2+i]
	for k in range(5):
		dataRefContent[j][k] = dataQuestion1[3+j*4][10+k]
		dataRefPurpose[j][k] = dataQuestion1[3+j*4][15+k]
labelAction = [0 for i in range(8)]
for i in range(8):
	labelAction[i] = dataQuestion1.dtype.names[2+i]
labelRefContent = [0 for j in range(5)]
labelRefPurpose = [0 for j in range(5)]
for j in range(5):
	labelRefContent[j] = dataQuestion1.dtype.names[10+j]
	labelRefPurpose[j] = dataQuestion1.dtype.names[15+j]

plt.subplot(amountPlot, 1, counterPlot)
plt.barh(np.arange(8)*2, dataAction[0], 0.6, color=colorF2F, label="F2F")
plt.barh(np.arange(8)*2+0.6, dataAction[1], 0.6, color=colorCSCW, label="CSCW")
plt.yticks(np.arange(8)*2, labelAction)
plt.xlabel('Amount of occurrence')
plt.ylabel('Action')
plt.title('Action frequency distribution')
plt.legend(('F2F', 'CSCW'))
plt.show()
counterPlot += 1

plt.subplot(amountPlot, 1, counterPlot)
plt.barh(np.arange(5)*2, dataRefContent[0], 0.6, color=colorF2F, label="F2F")
plt.barh(np.arange(5)*2+0.6, dataRefContent[1], 0.6, color=colorCSCW, label="CSCW")
plt.yticks(np.arange(5)*2, labelRefContent)
plt.xlabel('Amount of occurrence')
plt.ylabel('Reference Content')
plt.title('Reference content frequency distribution')
plt.legend(('F2F', 'CSCW'))
plt.show()
counterPlot += 1

plt.subplot(amountPlot, 1, counterPlot)
plt.barh(np.arange(5)*2, dataRefPurpose[0], 0.6, color=colorF2F, label="F2F")
plt.barh(np.arange(5)*2+0.6, dataRefPurpose[1], 0.6, color=colorCSCW, label="CSCW")
plt.yticks(np.arange(5)*2, labelRefPurpose)
plt.xlabel('Amount of occurrence')
plt.ylabel('Reference Purpose')
plt.title('Reference purpose frequency distribution')
plt.legend(('F2F', 'CSCW'))
plt.show()
counterPlot += 1
```
