```python
import numpy as np
%matplotlib inline
import matplotlib
matplotlib.rcParams.update({'font.size':14})
import matplotlib.pyplot as plt 
counterPlot = 1
amountPlot = 6
plt.rcParams["figure.figsize"]=[10,7*amountPlot]
colorF2F = '#8BA6FF'
colorCSCW = '#CC6C52'
colorAction0 = ['#A2BAFF', '#7FA3FF', '#7293E5', '#5F7ABF', '#6D9AFF', '#40527F', '#222A40', '#202940']
colorAction1 = ['red', 'blue', 'yellow', 'green', 'black', 'grey', 'yellow', 'green']


dataQuestion1 = np.genfromtxt('Question1.csv', dtype=int, delimiter=',', skip_header=1,
	names=True)
dataAction = [[[0 for i in range(8)] for j in range(4)] for k in range(2)]
dataRefContent = [[[0 for ii in range(5)] for j in range(4)] for k in range(2)]
dataRefPurpose = [[[0 for ii in range(5)] for j in range(4)] for k in range(2)]
# dataXXX[0][][]: F2F; dataXXX[1][][]: CSCW
# dataX[X[F2F/CSCW][Task No.][Action/Ref Content/Ref Purpose]
for k in range(2):
	for j in range(4):
		for i in range(8):
			dataAction[k][j][i] = dataQuestion1[j+k*4][2+i]
		for ii in range(5):
			dataRefContent[k][j][ii] = dataQuestion1[j+k*4][10+ii]
			dataRefPurpose[k][j][ii] = dataQuestion1[j+k*4][15+ii]
labelAction = [0 for i in range(8)]
for i in range(8):
	labelAction[i] = dataQuestion1.dtype.names[2+i]
labelRefContent = [0 for j in range(5)]
labelRefPurpose = [0 for j in range(5)]
for j in range(5):
	labelRefContent[j] = dataQuestion1.dtype.names[10+j]
	labelRefPurpose[j] = dataQuestion1.dtype.names[15+j]

# question 1
plt.subplot(amountPlot, 1, counterPlot)
plt.barh(np.arange(8)*2, dataAction[0][3], 0.6, color=colorF2F, label="F2F")
plt.barh(np.arange(8)*2+0.6, dataAction[1][3], 0.6, color=colorCSCW, label="CSCW")
plt.yticks(np.arange(8)*2+0.3, labelAction)
plt.xlabel('Amount of occurrence')
plt.ylabel('Action mode')
plt.title('Action mode frequency distribution')
plt.legend(('F2F', 'CSCW'))
plt.show()
counterPlot += 1

plt.subplot(amountPlot, 1, counterPlot)
plt.barh(np.arange(5)*2, dataRefContent[0][3], 0.6, color=colorF2F, label="F2F")
plt.barh(np.arange(5)*2+0.6, dataRefContent[1][3], 0.6, color=colorCSCW, label="CSCW")
plt.yticks(np.arange(5)*2+0.3, labelRefContent)
plt.xlabel('Amount of occurrence')
plt.ylabel('Deixis reference')
plt.title('Deixis reference content frequency distribution')
plt.legend(('F2F', 'CSCW'))
plt.show()
counterPlot += 1

plt.subplot(amountPlot, 1, counterPlot)
plt.barh(np.arange(5)*2, dataRefPurpose[0][3], 0.6, color=colorF2F, label="F2F")
plt.barh(np.arange(5)*2+0.6, dataRefPurpose[1][3], 0.6, color=colorCSCW, label="CSCW")
plt.yticks(np.arange(5)*2+0.3, labelRefPurpose)
plt.xlabel('Amount of occurrence')
plt.ylabel('Deixis role')
plt.title('Deixis role frequency distribution')
plt.legend(('F2F', 'CSCW'))
plt.show()
counterPlot += 1

# # question 2
# # Task Vs. Action
# labelTask = ['Task 3', 'Task 2', 'Task 1']
# for i in range(2):
# 	del dataAction[i][-1]
# dataActionPerTask = np.flip(np.transpose(np.reshape(dataAction,(6, 8))), 1)
# leftValue0 = [0,0,0]
# leftValue1 = [0,0,0]
# plt.subplot(amountPlot, 1, counterPlot)
# for i in range(8):
# 	plt.barh(np.arange(3)*2, dataActionPerTask[i][3:6], 0.6, left=leftValue1, color=colorAction0[i], label=labelAction[i])	
# 	plt.barh(np.arange(3)*2+0.6, dataActionPerTask[i][0:3], 0.6, left=leftValue0, color=colorAction1[i], label=labelAction[i])	
# 	leftValue0 += dataActionPerTask[i][3:6]
# 	leftValue1 += dataActionPerTask[i][0:3]
# plt.yticks(np.arange(3)*2+0.3, labelTask)
# plt.xlabel('Occurrence amount of each action mode')
# plt.ylabel('Task')
# plt.show()
```
